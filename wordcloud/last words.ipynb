```python
from wordcloud import WordCloud
```

```python


# Read the whole text.
text = open('lastwords.txt').read()

# Generate a word cloud image (different each time)
wordcloud = WordCloud(font_path="corbel.ttf", width=1000, height=1000,max_words=400,background_color = "black", colormap = "coolwarm").generate(text)

# Display the generated image:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()

wordcloud.to_file("cloud.png")
```
